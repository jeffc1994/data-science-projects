```python
import pandas as pd
import os

# Folder containing .DAT files
dat_folder = r"C:\Users\JEFF\data-science-projects\housing_data_project\DAT files"

# List to store extracted rows
b_rows = []

# Parse all .DAT files in the folder
for filename in os.listdir(dat_folder):
    if filename.lower().endswith(".dat"):
        with open(os.path.join(dat_folder, filename), "r", encoding="utf-8") as f:
            for line in f:
                if line.startswith("B;"):
                    parts = line.strip().split(";")
                    b_rows.append(parts)

# Build DataFrame
b_df = pd.DataFrame(b_rows)
```

```python
b_df.columns = [
    "record_type",               # 'B'
    "district_code",             # 3B-2
    "property_id",               # 3B-3
    "sale_counter",              # 3B-4
    "lodgement_datetime",       # 3B-5
    "property_name",            # 3B-6
    "unit_number",              # 3B-7
    "house_number",             # 3B-8
    "street_name",              # 3B-9
    "locality",                 # 3B-10
    "postcode",                 # 3B-11
    "land_area",                # 3B-12
    "area_type",                # 3B-13
    "contract_date",            # 3B-14
    "settlement_date",          # 3B-15
    "purchase_price",           # 3B-16
    "zoning",                   # 3B-17
    "nature_of_property",       # 3B-18
    "primary_purpose",          # 3B-19
    "strata_lot_number",        # 3B-20
    "component_code",           # 3B-21
    "sale_code",                # 3B-22
    "sale_interest_percent",    # 3B-23
    "dealing_number",            # 3B-24
    "trailing_empty"]
```

```python
df = b_df.drop(columns=['trailing_empty'])
```
